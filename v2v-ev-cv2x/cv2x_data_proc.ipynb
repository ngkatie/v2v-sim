```python
import pandas as pd
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt

dataset = pd.read_csv('ev_cumulative.csv')
cv2x = pd.DataFrame(dataset)

matplotlib.style.use('ggplot')
```

```python
nod40 = cv2x.query("num_nodes == 40")
sns.lineplot(data=nod40,x="t2",y="avg_latency_ms",label="40 nod")

nod35 = cv2x.query("num_nodes == 35")
sns.lineplot(data=nod35,x="t2",y="avg_latency_ms",label="35 nod")

nod30 = cv2x.query("num_nodes == 30")
sns.lineplot(data=nod30,x="t2",y="avg_latency_ms",label="30 nod")

nod25 = cv2x.query("num_nodes == 25")
sns.lineplot(data=nod25,x="t2",y="avg_latency_ms",label="25 nod")

nod20 = cv2x.query("num_nodes == 20")
sns.lineplot(data=nod20,x="t2",y="avg_latency_ms",label="20 nod")

plt.title("Selection Window Length and Average Latency")
plt.xlabel("Selection Window End Time (ms)")
plt.ylabel("Average Latency (ms)")
plt.legend()

plt.savefig("cv2x_latency_plot")
```
